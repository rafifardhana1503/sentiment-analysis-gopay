# Import Library

```python
from google_play_scraper import reviews_all, Sort
import pandas as pd
```

# Scraping Data

```python
# Scraping ulasan aplikasi GoPay di Google Play Store
app_id = "com.gojek.gopay"

scrapreview = reviews_all(
    app_id,
    lang="id",
    country="id",
    sort=Sort.NEWEST
)
```

```python
# Konversi ke DataFrame
df = pd.DataFrame(scrapreview)
df.head()
```

```python
# Simpan ke file CSV
df.to_csv("dataset_gopay_reviews.csv", index=False)
```
